```python
import os
import folium
import pandas as pd

print(folium.__version__)
```

```python
import geopandas as gpd
data1_2008 = pd.read_excel("Directionality Coding 2008-2016.xlsx", sheet_name="2008")
data1_2009 = pd.read_excel("Directionality Coding 2008-2016.xlsx", sheet_name="2009 RECODE")
data1_2010 = pd.read_excel("Directionality Coding 2008-2016.xlsx", sheet_name="2010")
```

```python
data1_2008 = data1_2008.rename(columns={2008:'Years'})
data1_2009 = data1_2009.rename(columns={2009:'Years'})
data1_2010 = data1_2010.rename(columns={2010:'Years'})
```

```python
frame=[data1_2008,data1_2009,data1_2010]
gdf=pd.concat(frame)
```

```python
gdf.head()
```

```python
from geopy import geocoders
from geopy.geocoders import Nominatim
gn = geocoders.GeoNames(username='rose0037')
geolocator = Nominatim()
```

```python
for country in gdf['Origin Country'].unique():
    print(country)
    if (country == "Sub-Saharan Africa"):
        location = geolocator.geocode("Africa")
        
    else:
        if (country == "Former Soviet states" or country == "Former Soviet Union" ):
            location = geolocator.geocode("Soviet union")
        else:
            if (country != "Neighboring countries"):
                location = geolocator.geocode(country)
    gdf.loc[(gdf['Origin Country']==country), 'Source_Latitude'] = location.latitude
    gdf.loc[(gdf['Origin Country']==country), 'Source_Longitude'] = location.longitude
```

```python

import numpy as np


styledata1 = {}

for country in gdf1.index:
    df1 = pd.DataFrame(
        {'color': np.random.normal(size=n_periods),
         'opacity': np.random.normal(size=n_periods)},
        index=dt_index1
    )
    df1 = df1.cumsum()
    df1.sample(n_sample, replace=False).sort_index()
    styledata1[country] = df1
    
```

```python
import numpy as np


styledata = {}

for country in df_counters.index:
    df = pd.DataFrame(
        {'color': np.random.normal(size=len(df_counters.Years.unique())),
         'opacity':np.random.normal(size=len(df_counters.Years.unique()))},
        index=dt_index
    )
   
    df = df.cumsum()
    df.sample(len(df_counters.Years.unique()), replace=False).sort_index()
    styledata[country] = df
```

```python
max_color1, min_color1, max_opacity1, min_opacity1 = 0, 0, 0, 0

for country, data in styledata1.items():
    max_color1 = max(max_color1, data['color'].max())
    min_color1 = min(max_color1, data['color'].min())
    max_opacity1 = max(max_color1, data['opacity'].max())
    max_opacity1 = min(max_color1, data['opacity'].max())
```

```python
max_color, min_color, max_opacity, min_opacity = 0, 0, 0, 0

for country, data in styledata.items():
    max_color = max(max_color, data['color'].max())
    min_color = min(max_color, data['color'].min())
    max_opacity = max(max_color, data['opacity'].max())
    max_opacity = min(max_color, data['opacity'].max())
```

```python
from branca.colormap import linear


cmap1 = linear.PuRd_09.scale(min_color1, max_color1)

def norm(x):
    return (x - x.min()) / (x.max() - x.min())

for country, data in styledata1.items():
    data['color'] = data['color'].apply(cmap1)
    data['opacity'] = norm(data['opacity'])
```

```python
from branca.colormap import linear


cmap = linear.PuRd_09.scale(min_color, max_color)

def norm(x):
    return (x - x.min()) / (x.max() - x.min())

for country, data in styledata.items():
    data['color'] = data['color'].apply(cmap)
    data['opacity'] = norm(data['opacity'])
```
